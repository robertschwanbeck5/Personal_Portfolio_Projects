```python
import pandas as pd
from pandas.plotting import scatter_matrix

# these are some very standard math/science Python libraries
import scipy as sc
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 

# Scikit-learn is a widely used machine learning package
import sklearn as sk
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
#
# For starters, let's look at the way our dataset is constructed
# Let's look at the column headers, or "keys"
#
print('Keys of the iris dataset')
for k in dataset.keys():
    print('key = %s' % (k))
```

```python
#
# using the type() method, we can double check that
# our variable dataset is a data type called a
# scikit-learn Bunch, which is kind of like a Panda's data frame
#
type(dataset) 
```

```python
#
# Let's take a peek at the actual numeric data 
#
# note the way we access the dataset.  We specify which attribute
# we want ("data"), then specify which rows we want (here 1:5)
#
dataset['data'][1:5]
```

```python
#
# Let's look at some of the other parts of the data
#
# the "feature_names" corresponds to the column headers for the numeric data

dataset['feature_names']
```

```python
#
# Let's look at some of the other parts of the data
#
# the "target_names" corresponds to the known species for each row

dataset['target_names']
```

```python
#
# now let's visualize the data set 
# we will first create a Pandas data frame using the DataFrame() method
# then we'll use the scatter_matrix() method to make a graphic
#
pdata = pd.DataFrame( dataset['data'], columns=dataset.feature_names )
pd.plotting.scatter_matrix( pdata )
```

```python
#
# let's do some basic stats
# the data set has 50 observations per flower type
# so we can look at blocks of 50 rows at a time
#
for i in range(0,3):
    lo = 50*i
    hi = lo + 50
    name = dataset.target_names[i]
    print('data for flower type %s' % (name))
    print(pdata[lo:hi].describe())
```

```python
#
# let's do some basic visualizions
# the data set has 50 observations per flower type
# so we can look at blocks of 50 rows at a time
#

#
# make a list of blocks of data
#
blocks = []
for i in range(0,3):
    lo = 50*i
    hi = lo + 50
    block = pdata[lo:hi]
    blocks.append( block )

#
# first render the data as dots
#
plt.plot(blocks[0]['sepal length (cm)'], blocks[0]['sepal width (cm)'],'b.', label='setosa')
plt.plot(blocks[1]['sepal length (cm)'], blocks[1]['sepal width (cm)'],'r.', label='versicolor')
plt.plot(blocks[2]['sepal length (cm)'], blocks[2]['sepal width (cm)'],'g.', label='virginica')

# 
# set up plot embellishments
#
plt.ylim(0,5)
plt.xlim(0,10)
plt.title('Iris Data: Sepal Values')
plt.ylabel('sepal length (cm)')
plt.xlabel('sepal width (cm)')
plt.grid()
plt.legend(('setosa','versicolor','virginica'))
plt.show()
```
